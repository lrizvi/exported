```python
import pandas as pd, matplotlib.pyplot as plt

files = [
    'fsr_grasping.dmm.csv',
    '31#4_recapped_grasping.dmm.csv',
]

for f in files:
    df = pd.read_csv(f, header=[0], skiprows=[1], index_col=[0])
    df[df.Resistance < 1e30].plot.scatter(x='Time', y='Resistance', s=0.1)
    plt.savefig(f'{}.png', dpi=600, bbox_inches='tight')
```
